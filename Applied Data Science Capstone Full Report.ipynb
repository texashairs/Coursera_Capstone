# Best city in Louisiana for a new Fast Food Take-out Business in Light of Pandemic Restrictions

# Introduction / Business Problem


With the Covd-19 pandemic of 2020, the state of Louisiana has experienced a significant change in business along with the rest of the United States.  
Businesses such as bars and dine-in restaurants have curtailed business while firms such as big box retail and takeout restaurants have seen an increase in demand.  
With the changes in the business operating conditions, there is an opportunity for short term profitability with the right business in the right location.
Current restrictions in the state of Louisiana differ by parrish, but the following are common to all the large population centers in the state:
•	Bars are limited to 25% indoor capacity
•	Dine-in establishments are limited to 50% of seating capacity
•	Most entertainment venues (theaters, clubs) are closed or limited to 25% capacity
•	All current restrictions are found in https://ready.nola.gov/incident/coronavirus/safe-reopening/#377
Based on the current restrictions, take-out or fast-food restaurants are an opportunity to profit in the current business climate.  A venue such as a fast food restaurant 
or food truck could fill the gap created by limited bar and dine-in capacity.   

Therefore, a use of a venue database could be used to determine which areas of the state of Louisiana could be under-served by fast food restaurants 
where a food truck or small take-out restaurant could be profitable.

# Data

The data that will be sourced for this project will involve the following
•	https://www.louisiana-demographics.com/cities_by_population - a list population centers in the state of Louisiana which will be pulled into a pandas dataframe.
•	The Foursquare API database.  Venues will be called from the API based on the latitude and longitude of the top 8 population centers in Louisiana.
•	Current Covid-19 restrictions in the state of Louisiana for data interpretation (https://ready.nola.gov/incident/coronavirus/safe-reopening/#377)

The dataframe of the population will be run through geocoder to determine the latitude and longitude of each of the top 8 population centers in the state.  
This information will feed into a call to the Foursquare API to pull a list of venues 2500 meters from the population centers.  
From there the data will be analyzed and a recommendation made.

Data cleaning was minimal as I picked the Lousiana cities by population website due to it having minimal columns.  Reading the html into the data frame using the pd.read_html command resulted in a clean dataframe with city and population.  


# Methodology

The approach used for the analysis was very similar to the week 3 lab using k-means to classify clusters of venues in Manhattan neighborhoods.  I modified the analysis to compare the top 8 population centers in Louisiana.  

The first step after preparing the city dataframe was adding latitude and longitude using geocoder using a defined function, an idea I borrowed from https://www.geeksforgeeks.org/how-to-find-longitude-and-latitude-for-a-list-of-regions-or-country-using-python/.  It resulted in a nice dataframe that shows as output in line 10 of my code.

I built a folium map on this dataframe centered around LeCompte, LA as a center point of the state.  Alexandria, LA is more often considered the "center" of the state of Louisiana yet geocoder repeatedly tried to center the map in Alexandria, Egypt, instead!  As a result I used the lesser known town in central Louisiana to build my initial folium map, which does not translate well to Github.  Visually looking at the locations of the top 8 cities, they are located in 5 principal geographic areas.  The Northwest (NW) corner of Louisiana features two cities, the Southwest corner, south Central, Baton Rouge and the SE corner of Louisiana are the five apparent areas of clustering.

Next I used my code to pull venues from these 8 cities in Louisiana, using a 2500 meter radius on each.  Cities in Louisiana are not nearly as densely populated as Manhattan, so a larger radius was needed to pull a more representative sample of venues.  Line 15 in my code shows a groupby of the dataframe showing 44 to 100 venues pulled from the API call per city, indicating the data should be a representative sample for these top 8 cities. Line 61 determined that there were 172 unique venue categories.

Prior to using k-means, one-hot encoding was performed followed by grouping the venue data by city into a new dataframe, with the output showing in line 65.  The dataframe shows the top 10 venues by city, and visually I could already see cities such as Bossier City, LA were well represented in fast food and take-out dining in three of the top five venues.  The question would be how do the venues of the cities cluster together, and which city stands out as most under-represented for fast food and take-out dining?

K-means was run with a setting of 5 due to the apparent 5 geographic clusters of cities out of the top 8 by population.  After building a dataframe with the k-means clusters, I created a new folium map (again, will not show in Github) with the clusters all marked with a separate color.  I found it very interesting that cities close in geographic proximity (New Orleans, Kenner, Metairie in SE corner of map) were not clustered together as venues.  Instead, a city such as Metairie in the SE corner of Louisiana was clustered similar to Bossier City in NW Louisiana.  The K-means clustering analysis is very helpful in that regard to provide insights that a geographic analysis alone would overlook.

My final analysis was to group the top 10 venue dataframe by cluster to visually determine which city cluster shows the least representation in fast food - which is promoted during the pandemic - and the highest representation of bar and clubs that are restricted during the pandemic.  This final dataframe is displayed in the results section below and shows in line 28 of my code.

# Discussion
Cluster 0, consisting of Lafayette, LA and Lake Charles LA are represented with Fast Food and Fried Chicken in their top 5 venues, with Pizza also represented in the top 10.  Though the bars in each city will be restricted, Foursquare indicates there is already a significant presence of takeout options in this cluster, and would not be ideal for opening a new take-out venue or food truck.

Cluster 1 conists only of Kenner, LA.  Fast Food, Coffee and Pizza are in the top 5 venues, which again suggest this cluster is well served with take-out dining.

Cluster 3 consists of Metairie and Bossier City, LA.  This cluster is represented with take-out options (Fast food, Fried Chicken, Smoothie, Burger Joint) in the top 6 venues.

Cluster 4 consists of Baton Rouge and Shreveport, LA.  This cluster is reliant on bars in the top 2 venues but has fast food, burger joints and "American Restaurant" available in the top 10 venues to service needs for take-out dining.

Cluster 2 consists only of New Orleans, LA.  The top 4 venues are dine-in restaurants (Cajun, Seafood, Italian) and a Hotel, which are all affected by current pandemic restrictions.  The coffee shop at number 5 perhaps would function for take-out but the 6th through 10th venues are bars and music venues that are either closed or restricted in capacity in the current pandemic econoomy.

Based on the k-means clustering, my recommendation would be to invest in fast food or a food-truck venue for New Orleans, LA as an under-served area for take-out dining during the pandemic.  The restrictions in New Orleans with its large population will lead to demand for good quality take-out dining options until later in 2021 when restrictions can be relaxed.
