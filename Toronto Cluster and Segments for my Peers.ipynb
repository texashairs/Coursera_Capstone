```python
import matplotlib.pyplot as plt
from unicodedata import normalize

table_CA = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M', match='Postal Code')
# Got the idea for pd.read_html from https://pbpython.com/pandas-html-table.html, great tool!
```

```python
print(f'Total tables: {len(table_CA)}')
```

```python
dfCA= table_CA[0]
```

```python
dfCA.head(12)
```

```python
droprows = dfCA[dfCA['Borough']=='Not assigned'].index
# Goal to eliminate the rows of Not assigned boroughs
```

```python
dfCA.drop(droprows,inplace=True)
```

```python
dfCA.head(12)
#shows the Not assigned boroughs have been dropped from the dataframe.
```

```python
dfCA.loc[dfCA['Neighbourhood']=='Not assigned','Neighbourhood']=['Borough']
# Replacing any 'Not assigned' neighborhoods with the Borough column.
```

```python

dfCA['Postal_Code']=dfCA['Postal Code'].map(lambda x: x.lstrip('0123456789 '))
# this created a new Postal_Code column with the left strings removed!!!  YAY!!!!!!!!!!!
```

```python
dfCA.drop(['Postal Code'],axis=1)
```

```python
dfCA=dfCA[['Postal_Code','Borough','Neighbourhood']]
# Re-order the Postal_Code column to the 1st position
```

```python
dfCA.reset_index(drop=True, inplace=True)
# Drop the index column and reset index order.
```

```python
latlong = pd.read_csv('http://cocl.us/Geospatial_data')
# Pull in latitude + longitude of postal code data as geocoder won't work
```

```python
latlong.head()
# Let's see what we have to work with....
```

```python
dflatlong=latlong.set_index('Postal Code').join(dfCA.set_index('Postal_Code'))
```

```python
# The above code joined the dfCA dataframe to the latlong dataframe on the postal code column, making a 
# new dataframe called dflatlong
```

```python
dflatlong=dflatlong[['Borough','Neighbourhood','Latitude','Longitude']]
# Re-order the Latitude and Longitude to the right end of the dataframe
```

```python
#Start map around postal code M7A	Downtown Toronto	Queen's Park, Ontario Provincial Government	43.662301	-79.389494
latitude = 43.662301
longitude = -79.3894
neigh_map = folium.Map(location=[latitude, longitude], zoom_start=13)



# display map
neigh_map
```

```python
# add blue circle markers for each neighbuorhood in the various boroughs (Canadian spelling?)
for lat, lng, borough, neighborhood in zip(dflatlong['Latitude'], dflatlong['Longitude'], dflatlong['Borough'], dflatlong['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(neigh_map)  

# show map - this is the end of PART 3 of the assignment!
neigh_map
```
